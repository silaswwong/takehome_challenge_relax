```python
count.reset_index(level=0, inplace=True)
```

```python
count1 = count.drop(['index'],axis=1)
```

```python
count1.head()
```

```python
count2 = count1.drop(['time_stamp','visited'],axis=1)
```

```python
count2.head()
```

```python
count2['adopted_user'] = 1
count2.head()
```

```python
merge1 = count2.merge(users_df,left_on='user_id',right_on='object_id', how='right')
```

```python
merge1.head()
```

```python
merge1.info()
```

```python
merge2 = merge1.drop(['user_id','name','email','last_session_creation_time'],axis=1)
```

```python
merge2.head()
```

```python
merge3 = merge2.fillna(0)
```

```python
merge4 = merge3.sort_values('object_id').reset_index().drop(['index'],axis=1)
```

```python
merge4.head()
```

```python
merge4['invited_by_user_id'] = merge4['invited_by_user_id'].apply(lambda x : 1 if x>0 else 0)
```

```python
merge4.head()
```

```python
dummies = pd.get_dummies(merge4['creation_source']).drop(['SIGNUP_GOOGLE_AUTH'],axis=1)
```

```python
dummies.head()
```

```python
merge5 = pd.concat([dummies,merge4],axis=1).drop(['creation_source','creation_time','object_id','org_id'],axis=1)
```

```python
merge5
```

```python
independent = merge5.drop(['adopted_user'],axis=1)
dependent = merge5['adopted_user']
```

```python
X = independent.as_matrix()
y = dependent.as_matrix()
```
